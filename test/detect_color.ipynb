```python
import cv2
import numpy as np
from skimage import data
import matplotlib.pyplot as plt

#astro_cv = cv2.cvtColor(astro,cv2.COLOR_RGB2BGR)
astro_ori = data.astronaut()

plt.imshow(astro_ori)
```

```python
astro =astro_ori.copy()
astro.shape
astro[100:200,200:300] = (0,0,255)
plt.imshow(astro)
```

## [inRange()](https://docs.opencv.org/4.1.1/d2/de8/group__core__array.html#ga48af0ab51e36436c5d04340e036ce981)

`dst = cv.inRange(src, lowerb, upperb[, dst])`

### Parameters
* src	first input array.
* lowerb	inclusive lower boundary array or a scalar.
* upperb	inclusive upper boundary array or a scalar.
* dst	output array of the same size as src and CV_8U type. 

## [bitwise_and()](https://docs.opencv.org/4.1.1/d2/de8/group__core__array.html#ga60b4d04b251ba5eb1392c34425497e14)

`dst = cv.bitwise_and( src1, src2[, dst[, mask]])`

### Parameters
* src1 :first input array or a scalar.
* src2 :second input array or a scalar.
* dst  :output array that has the same size and type as the input arrays.
* mask :optional operation mask, 8-bit single channel array, that specifies elements of the output array to be changed. 

```python
lower_blue=(0,0,255)
upper_blue=(0,0,255)

img_mask   = cv2.inRange(astro, lower_blue, upper_blue)
img_result = cv2.bitwise_and(astro, astro, mask=img_mask)
```

```python
plt.imshow(img_result)

# convert image into gray scaled 
img_gray = cv2.cvtColor(img_result,cv2.COLOR_BGR2GRAY)


prj_along_y = np.sum(img_gray,0)
prj_along_x = np.sum(img_gray,1)


plt.figure()
plt.title('projection along x-axis')
plt.plot(prj_along_y)
plt.figure()
plt.title('projection along x-axis')
plt.plot(prj_along_x)
```
