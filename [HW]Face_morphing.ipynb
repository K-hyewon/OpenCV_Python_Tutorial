## Face morphing

```python
import cv2
import matplotlib.pyplot as plt

#이미지 준비하기
img0 = cv2.imread("MJ2017.jpg", cv2.IMREAD_UNCHANGED)
img1 = cv2.imread("MJ2020.jpg", cv2.IMREAD_UNCHANGED)

img_2017 = cv2.cvtColor(img0, cv2.COLOR_BGR2RGB)
img_2020 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

print(img_2017.shape)
print(img_2020.shape)

plt.figure()
plt.axis('off')
plt.imshow(img_2017)

plt.figure()
plt.axis('off')
plt.imshow(img_2020)
```

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

def f(x):
    alpha = x/100
    ret = cv2.addWeighted(img_2020, alpha, img_2017, 1-alpha, 0)
    plt.imshow(ret[:,:,::-1])

b, g, r = cv2.split(img_2017)
img_2017 = cv2.merge([r, g, b])

b, g, r = cv2.split(img_2020)
img_2020 = cv2.merge([r, g, b])

img_2020 = cv2.resize(img_2020, dsize=(img_2017.shape[1], img_2017.shape[0]), interpolation=cv2.INTER_AREA)

interact(f, x=widgets.IntSlider(min=0, max=100, step=10, value=10, continuous_update=False));
```

첫번재 쉘에서 이미지 준비할 때 bgr을 rgb로 바꿔줬는데 무슨 이유인지 두번째 쉘에서 머리와 옷을 제외하고 얼굴 부분이 다 파랗게 나와서 일단 이미지의 b, g, r을 각각 쪼개서 r, g, b로 재병합 해주었다. <br>
