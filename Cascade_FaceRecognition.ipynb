```
import numpy as np
import cv2
import matplotlib.pyplot as plt

face_cascade = cv2.CascadeClassifier('./haarcascade_frontalface_default.xml')

eye_cascade = cv2.CascadeClassifier('./haarcascade_eye.xml')

img = cv2.imread('./02_01.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(gray)

for (x,y,w,h) in faces:

  cv2.rectangle(img, (x,y), (x+w,y+h), (255,0,0) ,2 )
  roi = gray[y:y+h, x:x+w]
  eyes = eye_cascade.detectMultiScale(roi)

  cnt =0
  for (ex,ey,ew,eh) in eyes:
    if cnt <2:
      cv2.rectangle(img[y:y+h,x:x+w], (ex,ey), (ex+ew,ey+eh), (0,0,255), 2)
    else:
      pass
    cnt += 1

img = img[:,:,::-1]
plt.imshow(img)
plt.show()
```

```
import numpy as np
import cv2
import matplotlib.pyplot as plt

face_cascade_haar = cv2.CascadeClassifier('./haarcascade_profileface.xml')
face_cascade_lbp = cv2.CascadeClassifier('./lbpcascade_profileface.xml')


img = cv2.imread('./depositphotos_219491964-stock-photo-profile-face-young-woman-isolated.jpg')
#img = cv2.imread('./profile-picture-asian-2.jpg')
#img = img[:,280:,:]

#img = img[:,::-1,:]

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

faces_haar = face_cascade_haar.detectMultiScale(gray)
faces_lbp = face_cascade_lbp.detectMultiScale(gray)


print ('haar:',len(faces_haar))
for (x,y,w,h) in faces_haar:
  cv2.rectangle(img, (x,y), (x+w,y+h), (255,0,0) ,40 )

print ('lbp:',len(faces_lbp))
for (x,y,w,h) in faces_lbp:
  cv2.rectangle(img, (x,y), (x+w,y+h), (0,0,255) ,40 )

img = img[:,:,::-1]
plt.imshow(img)
plt.show()
```
