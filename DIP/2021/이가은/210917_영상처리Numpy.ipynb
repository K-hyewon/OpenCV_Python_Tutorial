# numpy.vsplit
## : numpy.vsplit(array, indices_or_sections) : Split an array into multiple sub-arrays vertically (row-wise)
## : axis 0로 array를 잘라 나눔

```python
import numpy as np
a = np.arange(12).reshape(-1,3)
a
```

```python
np.vsplit(a, (1,3))
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('/content/drive/MyDrive/가은/2021/영상처리/images/sample.jpeg')
subimgs = np.vsplit(img,4) #하나의 숫자로 주어질 경우 몇 등분인지를 나타냄
for c in subimgs:
  print(c.shape)
  plt.imshow(c[...,::-1])
  plt.show()
```

# numpy.split
## : numpy.split(array, indices_or_sections, axis=0) : Split an array into multiple sub-arrays as views into arrays.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/content/drive/MyDrive/가은/2021/영상처리/images/messi5.jpg')
subimgs = np.split(img,3,axis=0)
for idx,c in enumerate(subimgs):
  print(c.shape)
  plt.title(f'channel: {idx}')
  plt.imshow(c[...,::-1],cmap='gray')
  plt.show()
```

# numpy.where
## : numpy.where(condition[, x, y]) : Return elements chosen from x or y depending on condition.

## : condition : 검색에 사용될 조건.
## : x : condition이 True에 해당하는 위치에 지정되는 값 또는 array (검색에 사용된 ndarray로 broadcast처리 가능한 shape이어야 함)
## : y : condition이 False에 해당하는 위치에 지정되는 값 또는 array (검색에 사용된 ndarray로 broadcast처리 가능한 shape이어야 함)

```python
import numpy as np

a = np.arange(12).reshape(3,4)
x = np.array([[0 ,1,2,3],[10,11,12,13],[20,21,22,23]])
y = np.zeros((3,4))

a
```

```python
print(np.where(a > 5,x,y))
```

```python
import numpy as np
a = np.arange(12).reshape(3,4)

print(np.where(a > 5))
```

* 첫번째 ndarray는 axis 0에서 condition이 참인 위치들의 index임.
* 두번째 ndarray는 axis 1에서 condition이 참인 위치들의 index임.

```python
import numpy as np
a = np.arange(12).reshape(-1,4)
a
```

```python
coords = np.where( (a>3) & (a<8))
print(coords)
```

```python
c = np.zeros_like(a)
c[coords]=1
print(c)
```

# 4. NumPy : sum, mean, std, and so on

# numpy.sum : 총합을 구함
## : Sum of array elements over a given axis.
numpy.sum(a, axis=None, dtype=None, out=None, keepdims=<no value>, initial=<no value>, where=<no value>)

```python
import numpy as np
a = np.arange(20).reshape(5,-1)
sum = np.sum(a)
print(sum)
```

```python
import numpy as np
a = np.arange(20).reshape(5,-1)
row_sum_array = np.sum(a,axis=1)    # 행에서의 합

print(a)
print(row_sum_array.shape)
print(row_sum_array)
```

```python
import numpy as np
a = np.arange(20).reshape(5,-1)
row_sum_array = np.sum(a,axis=0)    # 열에서의 합

print(a)
print(row_sum_array.shape)
print(row_sum_array)
```

## : numpy.mean : 평균 구하기.
## : Compute the arithmetic mean along the specified axis.

numpy.mean(a, axis=None, dtype=None, out=None, keepdims=<no value>, *, where=<no value>)

```python
import numpy as np
a = np.arange(20).reshape(5,-1)
mean = np.mean(a)
print(mean)
```

```python
import numpy as np
a = np.arange(20).reshape(5,-1)
col_mean_array = np.mean(a,axis=0)    # 열에서 평균
print(a)
print(col_mean_array.shape )
print(col_mean_array )
```

```python
import numpy as np
a = np.arange(20).reshape(5,-1)
col_mean_array = np.mean(a,axis=1)    # 행에서 평균
print(a)
print(col_mean_array.shape )
print(col_mean_array )
```

## numpy.std : 표준편차 구하기.
## : Compute the standard deviation along the specified axis.

numpy.std(a, axis=None, dtype=None, out=None, ddof=0, keepdims=<no value>, *, where=<no value>)

```python
import numpy as np
a = np.arange(20).reshape(5,-1)
std = np.std(a)
print(std)
```

```python
import numpy as np
a = np.arange(20).reshape(5,-1)
col_std_array = np.std(a,axis=0)    # 열에서의 표준편차
print(a)
print(col_std_array.shape )
print(col_std_array )
```

```python
import numpy as np
a = np.arange(20).reshape(5,-1)
col_std_array = np.std(a,axis=1)    # 행에서의 표준편차
print(a)
print(col_std_array.shape )
print(col_std_array )
```
