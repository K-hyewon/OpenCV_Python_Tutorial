```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_img(path):
    img = cv2.imread(path, cv2.IMREAD_UNCHANGED)
    if img.shape[-1] == 4:
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGRA2RGBA))
    else:
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    return img
```

```python
sign = read_img('image/sign.jpg')
```

```python
sign = cv2.rotate(sign, cv2.ROTATE_90_CLOCKWISE)
```

```python
plt.imshow(cv2.cvtColor(sign, cv2.COLOR_RGB2BGR))
```

```python
def select_roi(img, save_path):
    x, y, w, h = cv2.selectROI('img', img, False)
    if w and h:
        print(f'x:{x: 3d}, y:{y: 3d}, w:{w: 3d}, h:{h: 3d}')
        roi = img[y:y+h, x:x+w]
        cv2.imshow('roi', roi)
        cv2.moveWindow('roi', 0,0)
        cv2.imwrite(save_path,roi)

    cv2.imshow('img',img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
select_roi(sign, 'image/roi.jpg')
```

```python
h, w, c = png_sign.shape
mask = png_sign[:,:,3]
mask_inv = cv2.bitwise_not(mask)

bgr_sign = png_sign[:,:,:3]
back_roi = background[:h, :w, :]

fg = cv2.bitwise_and(bgr_sign, bgr_sign, mask=mask)
bg = cv2.bitwise_and(back_roi, back_roi, mask=mask_inv)
add_roi = fg+bg

new_img = background.copy()
new_img[:h, :w, :] = add_roi

plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(fg, cv2.COLOR_BGR2RGB))
plt.title('fg')
plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(bg, cv2.COLOR_BGR2RGB))
plt.title('bg')
plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(add_roi, cv2.COLOR_BGR2RGB))
plt.title('add_roi')

plt.figure()
plt.imshow(cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB))
plt.title('new_img')
plt.show()
```

```python
def merge_img(fg_img, bg_img, save_path, where="left_top"):
    h, w, c = fg_img.shape
    mask = fg_img[:,:,3]
    mask_inv = cv2.bitwise_not(mask)

    bgr_img = fg_img[:,:,:3]
    fg = cv2.bitwise_and(bgr_img, bgr_img, mask=mask)

    new_img = bg_img.copy()

    if where=="left_top":
        back_roi = bg_img[:h, :w, :]
        bg = cv2.bitwise_and(back_roi, back_roi, mask=mask_inv)
        add_roi = fg+bg
        new_img[:h, :w, :] = add_roi
    elif where=="right_top":
        back_roi = bg_img[:h, new_img.shape[1]-w:, :]
        bg = cv2.bitwise_and(back_roi, back_roi, mask=mask_inv)
        add_roi = fg+bg
        new_img[:h, new_img.shape[1]-w:, :] = add_roi
    elif where=="left_bottom":
        back_roi = bg_img[new_img.shape[0]-h:, :w, :]
        bg = cv2.bitwise_and(back_roi, back_roi, mask=mask_inv)
        add_roi = fg+bg
        new_img[new_img.shape[0]-h:, :w, :] = add_roi
    else:
        back_roi = bg_img[new_img.shape[0]-h:, new_img.shape[1]-w:, :]
        bg = cv2.bitwise_and(back_roi, back_roi, mask=mask_inv)
        add_roi = fg+bg
        new_img[new_img.shape[0]-h:, new_img.shape[1]-w:, :] = add_roi

    plt.subplot(1,3,1)
    plt.imshow(cv2.cvtColor(fg, cv2.COLOR_BGR2RGB))
    plt.title('fg')
    plt.subplot(1,3,2)
    plt.imshow(cv2.cvtColor(bg, cv2.COLOR_BGR2RGB))
    plt.title('bg')
    plt.subplot(1,3,3)
    plt.imshow(cv2.cvtColor(add_roi, cv2.COLOR_BGR2RGB))
    plt.title('add_roi')
    
    plt.figure()
    plt.imshow(cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB))
    plt.title('new_img')
    plt.show()

    cv2.imwrite(save_path, new_img)
```

```python
merge_img(png_sign, background, 'image/merge.jpg', where="right_top")
```
