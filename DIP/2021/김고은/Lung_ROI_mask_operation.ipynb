```python
import cv2
import matplotlib.pyplot as plt

lung_roi = cv2.imread('/content/drive/MyDrive/Classroom/3-2/ImageProcessing/Images/Lung-ROI-segmentation.png')

rgb2bgr = cv2.cvtColor(lung_roi, cv2.COLOR_BGR2RGB)

plt.imshow(rgb2bgr)
plt.show()

lung_roi.shape 
```

```python
# resize (279,849,3)

re_lung_roi = cv2.resize(lung_roi, (849,279))
re_lung_roi.shape
```

```python
import numpy as np

subimgs = np.hsplit(re_lung_roi, 3)
plt.imshow(subimgs[0])
plt.show()
plt.imshow(subimgs[1])
plt.show()

img1 = subimgs[0]
img2 = subimgs[1]
```

```python
masked_roi = cv2.bitwise_and(img1, img2)

plt.figure()
plt.imshow(masked_roi,cmap=plt.cm.gray)
plt.show()
```
