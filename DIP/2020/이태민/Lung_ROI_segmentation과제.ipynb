shape함수를 사용하여 column의 수를 알아낸 다음 Mask 이미지와 X-ray이미지를 분리했다

```
s_image1=g_image[:,:283]
s_image2=g_image[:,284:567]

plt.figure()
plt.imshow(s_image1,cmap=plt.cm.gray)
plt.show()

plt.figure()
plt.imshow(s_image2,cmap=plt.cm.gray)
plt.show()
```

* 나눠진 부분의 영역을 확인

* bitwise_and()함수를 사용하여 ROI부분만을 남긴 이미지를 만들었다.

```
result_roi = cv2.bitwise_and(s_image1,s_image2)

plt.figure()
plt.imshow(result_roi,cmap=plt.cm.gray)
plt.show()
```
