# bitwise
## image load and split

```
import numpy as np
import matplotlib.pyplot as plt
import cv2

raw = cv2.imread('/content/drive/My Drive/DIP/Lung-ROI-segmentation.png',cv2.IMREAD_GRAYSCALE)

img  = raw[:,:int(raw.shape[1]/3)]
mask = raw[:,int(raw.shape[1]/3):int(2*raw.shape[1]/3)]

imgs = {'raw':raw, 'img':img, 'mask':mask}

plt.figure(figsize=[16,8])
for idx, (title, pic) in enumerate(imgs.items()):
  plt.subplot(1,3,idx+1)

  print(title,'shape : ',pic.shape)
  plt.title(title)
  plt.axis('off')
  plt.imshow(pic,cmap=plt.cm.gray)
plt.tight_layout()
plt.show()
```

## Masked ROI

```
plt.figure()
plt.imshow(img,cmap=plt.cm.gray)
plt.figure()
plt.imshow(mask,cmap=plt.cm.gray)
print(img.shape)
print(mask.shape)

#_, mask = cv2.threshold(mask,254,255,cv2.THRESH_BINARY)
masked_roi = cv2.bitwise_and(img,mask)


plt.figure()
plt.imshow(masked_roi,cmap=plt.cm.gray)
plt.show()
```
