```
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive/')
```

```
img = cv2.imread('/content/drive/My Drive/풍경사진.jpg', cv2.IMREAD_UNCHANGED)
img_1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_1)
```

BGR, RGB color 모델은 intensity를 변경할수없어서 YCrCb color모델로 변경 후 첫번째 채널에 있는 intensity만 equalization을 한 후 다시 RGB color모델로 변경하고 확인했다.

```
img_YCbCr = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
```

```
a, b, c = cv2.split(img_YCbCr)
```

```
a.shape
```

```
equalization = cv2.equalizeHist(a)
equalization
```

```
img_equal = cv2.merge((equalization, b, c))
img_equal.shape
```

```
img_equal = cv2.cvtColor(img_equal, cv2.COLOR_YCrCb2RGB)
```

```
plt.imshow(img_1)
```

```
plt.imshow(img_equal)
```

```
writefiles = cv2.cvtColor(img_equal, cv2.COLOR_RGB2BGR)
```

```
cv2.imwrite('equalization_풍경사진.jpg', writefiles)
```
