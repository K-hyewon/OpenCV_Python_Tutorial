```
import numpy as np
import cv2
from skimage import data
```

```
cat_rgb = data.chelsea() # take the test image of cat!
cat_gray = cv2.cvtColor(cat_rgb,cv2.COLOR_BGR2GRAY)

print(f"ndim:{cat_rgb.ndim},shape:{cat_rgb.shape},dtype:{cat_rgb.dtype},itemsize:{cat_rgb.itemsize}")
print(f"ndim:{cat_gray.ndim},shape:{cat_gray.shape},dtype:{cat_gray.dtype},itemsize:{cat_gray.itemsize}")
```

```
import matplotlib.pyplot as plt

profile = cat_gray[40,:]
plt.plot(profile)
plt.show()
```

```
x = np.arange(10)
y = x**2 +3
plt.plot(x,y)
plt.show()
```

# Color character
* b : blue
* g : green
* r : red
* c : cyan
* m : megenta
* y : yellow
* k : black
* w : white

```
plt.plot(x,y,'m')
plt.show()
```

```
plt.plot(x,y,'c')
plt.show()
```

# Style

* '-' : line
* '.' : dot
* '-.' : line dot
* 'o' : circle
* '^' : triangle
* 'v' : reverse triangle
* '<' : left triangle
* '>' : right triange
* 'h' : hexagon
* 'D' : diamond
* 'x' : x
* 's' : square
* '*' : star
* ',' : pixel

```
f1 = x*5
f2 = x*10
f3 = x**2

plt.plot(x,f1,'k,')
plt.plot(x,f2,'r--')
plt.plot(x,f3,'mh')
plt.show()
```

```
plt.subplot(1,3,1)
plt.plot(x,f1,'k-')

plt.subplot(1,3,2)
plt.plot(x,f2,'ro')

plt.subplot(1,3,3)
plt.plot(x,f2,'bx')

plt.show()
```
