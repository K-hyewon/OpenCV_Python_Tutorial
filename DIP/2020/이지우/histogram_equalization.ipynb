```
import cv2
import os

if os.path.isfile('/content/drive/My Drive/2020-2-자료/DIP/IMG_5030.JPG'): 
  flower = cv2.imread('/content/drive/My Drive/2020-2-자료/DIP/IMG_5030.JPG')
else:
  print('no file')

rgb_flower = flower[:,:,::-1]
```

```
import matplotlib.pyplot as plt
plt.imshow(rgb_flower)
plt.show()
```

이 이미지를 histogram equalization 시키자

```
r,g,b = cv2.split(rgb_flower)
r_new = cv2.equalizeHist(r)
g_new = cv2.equalizeHist(g)
b_new = cv2.equalizeHist(b)

channel_new = cv2.merge([r_new,g_new,b_new])

plt.imshow(channel_new)
```

채널별로 histogram equalization 시킨 이미지를 확인할 수 있다

```
YCrCb_flower = cv2.cvtColor(flower, cv2.COLOR_BGR2YCrCb)

Y, Cr, Cb = cv2.split(YCrCb_flower)
Y_new = cv2.equalizeHist(Y) #밝기 영역 Y 채널을 평준화

YCrCb_new = cv2.merge([Y_new,Cr,Cb]) # Cr, Cb 채널과 병합
yCrCb_new_rgb = cv2.cvtColor(YCrCb_new, cv2.COLOR_YCrCb2BGR)
yCrCb_new_rgb = yCrCb_new_rgb[:,:,::-1]

plt.figure(figsize=(10,10))
plt.subplot(2,2,1),plt.imshow(rgb_flower)
plt.subplot(2,2,2),plt.imshow(yCrCb_new_rgb)
plt.show()
```

YCbCr로 histogram equalizationn 시킨 이미지를 확인할 수 있다

```
YCrCb_flower = cv2.cvtColor(flower, cv2.COLOR_BGR2YCrCb)

Y, Cr, Cb = cv2.split(YCrCb_flower)
Cr_new = cv2.equalizeHist(Cr) #Cr채널을 평준화

YCrCb_new2 = cv2.merge([Y,Cr_new,Cb]) # Y, Cb 채널과 병합
yCrCb_new_rgb2 = cv2.cvtColor(YCrCb_new2, cv2.COLOR_YCrCb2BGR)
yCrCb_new_rgb2 = yCrCb_new_rgb[:,:,::-1]

plt.figure(figsize=(10,10))
plt.subplot(2,2,1),plt.imshow(rgb_flower)
plt.subplot(2,2,2),plt.imshow(yCrCb_new_rgb2)
plt.show()
```

평준화시킨 채널의 차이를 확인할 수 있다

(Y채널과 Cr채널)

```
HSV_flower = cv2.cvtColor(flower, cv2.COLOR_BGR2HSV)

H, S, V = cv2.split(HSV_flower)
V_new = cv2.equalizeHist(V) #Value 영역인 V채널 평준화

HSV_new = cv2.merge([H, S, V_new]) #병합
HSV_new_rgb = cv2.cvtColor(HSV_new, cv2.COLOR_HSV2BGR)
HSV_new_rgb = HSV_new_rgb[:,:,::-1]

plt.figure(figsize=(10,10))
plt.subplot(2,2,1),plt.imshow(rgb_flower)
plt.subplot(2,2,2),plt.imshow(HSV_new_rgb)
plt.show()
```

HSV로 histogram equalization 시킨 이미지를 확인할 수 있다

```
HSV_flower = cv2.cvtColor(flower, cv2.COLOR_BGR2HSV)

H, S, V = cv2.split(HSV_flower)
H_new = cv2.equalizeHist(H) #H채널 평준화

HSV_new2 = cv2.merge([H_new, S, V]) #병합
HSV_new_rgb2 = cv2.cvtColor(HSV_new2, cv2.COLOR_HSV2BGR)
HSV_new_rgb2 = HSV_new_rgb[:,:,::-1]

plt.figure(figsize=(10,10))
plt.subplot(2,2,1),plt.imshow(rgb_flower)
plt.subplot(2,2,2),plt.imshow(HSV_new_rgb2)
plt.show()
```
