## Load Image

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

#이미지 파일을 Grayscale로 읽어옴
img = cv2.imread("Lung-ROI-segmentation.png", cv2.IMREAD_GRAYSCALE)
```

```python
#이미지 확인
plt.axis('off')
plt.imshow(img,cmap=plt.cm.gray)
plt.show()
```

```python
#이미지 크기 확인
img.shape
```

## Split Image

그림판으로 너비를 확인한 결과 미세한 움직임에 따라 1-2 정도가 차이 나서, 이미지 width를 3등분해서 쪼갬

```python
raw_image = img.copy() #잘못 자르면 돌려야하므로 이미지 카피본 사용

xray = raw_image[:,:int(raw_image.shape[1]/3)]
mask = raw_image[:,int(raw_image.shape[1]/3):int(2*raw_image.shape[1]/3)]
```

```python
print('xray 사이즈: ', xray.shape)
plt.axis('off')
plt.imshow(xray, cmap=plt.cm.gray)
plt.show()
```

```python
print('mask 사이즈: ', mask.shape)
plt.axis('off')
plt.imshow(mask, cmap=plt.cm.gray)
plt.show()
```

## Add Image - Masking ROI

```python
ROI_img = cv2.bitwise_and(xray, mask)

plt.axis('off')
plt.imshow(ROI_img, cmap=plt.cm.gray)
plt.show()
```
