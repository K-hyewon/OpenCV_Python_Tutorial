```
import cv2
import numpy as np
import argparse
import os
import matplotlib.pyplot as plt
```

```
from google.colab.patches import cv2_imshow
```

```
path = '/content/drive/My Drive/dobby.jpg' 

# 입력 받은 이미지를 불러옵니다.
src = cv2.imread(path)

#1. hsv 컬러 형태로 변형
hsv = cv2.cvtColor(src, cv2.COLOR_BGR2HSV)
# h, s, v로 컬러 영상을 분리
h, s, v = cv2.split(hsv)
# v값 히스토그램 평활화
equalizedV = cv2.equalizeHist(v)
# h,s,equalizedV를 합쳐서 새로운 hsv 이미지로
hsv2 = cv2.merge([h,s,equalizedV])
# 마지막으로 hsv2를 다시 BGR 형태로 변경
hsvDst = cv2.cvtColor(hsv2, cv2.COLOR_HSV2BGR)
#출력하여 비교해보기
cv2_imshow(src)
cv2_imshow(hsvDst)
```

```
#2. YCrCb 컬러 형태로 변환

yCrCb = cv2.cvtColor(src, cv2.COLOR_BGR2YCrCb)
# y, Cr, Cb로 컬러 영상을 분리
y, Cr, Cb = cv2.split(yCrCb)
# y값을 히스토그램 평활화.
equalizedY = cv2.equalizeHist(y)
# equalizedY, Cr, Cb를 합쳐서 새로운 yCrCb 이미지로
yCrCb2 = cv2.merge([equalizedY, Cr, Cb])
# 마지막으로 yCrCb2를 다시 BGR 형태로 변경
yCrCbDst = cv2.cvtColor(yCrCb2, cv2.COLOR_YCrCb2BGR)
#출력
cv2_imshow(src)
cv2_imshow(hsvDst)
```

```
import matplotlib.pyplot as plt
```

```
#3개 비교 #PLT는 RGB라 BGR->RGB
src_RGB = cv2.cvtColor(src, cv2.COLOR_BGR2RGB)
hsvDst_RGB = cv2.cvtColor(hsv2, cv2.COLOR_HSV2RGB)
yCrCbDst_RGB = cv2.cvtColor(yCrCb2, cv2.COLOR_YCrCb2RGB)
plt.figure(figsize=(20,20))
plt.subplot(1,3,1)
plt.imshow(src_RGB)
plt.subplot(1,3,2)
plt.imshow(hsvDst_RGB)
plt.subplot(1,3,3)
plt.imshow(yCrCbDst_RGB)
plt.show()
```
