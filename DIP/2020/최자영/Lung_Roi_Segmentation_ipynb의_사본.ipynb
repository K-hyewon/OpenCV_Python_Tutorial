```
import numpy as np
import cv2
```

```
#Grayscale로 먼저 불러옴
img_path = '/content/drive/My Drive/Lung-ROI-segmentation.png'
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE) 
```

```
from google.colab.patches import cv2_imshow
cv2_imshow(img)
```

```
np.shape(img)
```

```
num1 = int(np.shape(img)[1]/3)
```

```
img_cp = img.copy()
img_cp.astype
```

```

#슬라이싱함
from google.colab.patches import cv2_imshow
img_split1 = img_cp[:, :num1]
cv2_imshow(img_split1)
```

```
from google.colab.patches import cv2_imshow
img_split2 = img_cp[:,num1:2 * num1]
cv2_imshow(img_split2)
```

```
from google.colab.patches import cv2_imshow
img_split3 = img_cp[:,2 * num1:3 * num1]
cv2_imshow(img_split3)
```

```
# 폐영역 검출
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
plt.subplot(1,3,1)
plt.imshow(img_split1, cmap='gray')
plt.subplot(1,3,2)
plt.imshow(img_split2, cmap='gray')
plt.subplot(1,3,3)
plt.imshow(img_split3, cmap='gray')
plt.imshow(img_split2, cmap='Reds', alpha=0.5)
plt.show()
```

```
# 폐영역 검출
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
plt.subplot(1,3,1)
plt.imshow(img_split1, cmap='gray')
plt.subplot(1,3,2)
plt.imshow(img_split2, cmap='gray')
plt.subplot(1,3,3)
plt.imshow(cv2.bitwise_and(img_split1,img_split2), cmap='gray')
plt.show()
```
