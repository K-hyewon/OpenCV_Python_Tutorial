```
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt

#구글 드라이브에서 이미지 불러오기
path = '/content/drive/My Drive/2020-2_Image Processing'
fstr_0 = os.path.join(path,'YL_2019.jpg')
fstr_1 = os.path.join(path,'YL_2003.jpg')

img_2019  = cv2.imread(fstr_0)
img_2019_rgb = cv2.cvtColor(img_0,cv2.COLOR_BGR2RGB)

img_2003  = cv2.imread(fstr_1)
img_2003_rgb = cv2.cvtColor(img_1,cv2.COLOR_BGR2RGB)

print(img_2019_rgb.shape)
print(img_2003_rgb.shape)

plt.figure()
plt.imshow(img_2019_rgb, interpolation='bicubic')
plt.show()

plt.figure()
plt.imshow(img_2003_rgb, interpolation='bicubic')
plt.show()
```

```
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

def f(x):
  alpha = x/100
  ret = cv2.addWeighted(img_2019, alpha, img_2003, 1-alpha,0)
  plt.imshow(ret[:,:,::-1], interpolation='bicubic')

img_2019 = cv2.resize(img_2019, dsize=(img_2003.shape[1],img_2003.shape[0]), interpolation= cv2.INTER_AREA)

interact(f, x=widgets.IntSlider(min=0, max=100, step=10, value=10, continuous_update=False));
```
